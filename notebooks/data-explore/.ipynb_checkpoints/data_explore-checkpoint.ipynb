```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('../data/archive/DR3_6D_kinematics_2000.csv')
data.columns
```

```python
sns.kdeplot(data=data, x='')
```

```python
sns.histplot(data=data, x='feH')
```

```python
sns.histplot(data=data, x='x') 
# x-coordinate is skewed negative because the center is the sun
```

```python
sns.histplot(data=data, x='r')
# again the skew is because of where sun is
```

```python
sns.histplot(data=data, x='r_helio')
# this makes no sense
```

```python
sns.histplot(data=data, x='feH') 
print(data['feH'])
# positive direction is the direction of sun's movement
```

```python
sns.histplot(data=data, x='vphi') # clockwise rotation
```

```python
sns.scatterplot(data=data, x='vphi', y='y', hue='feH', size=0.01) 
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=5)
```

```python
data_pca = data[['vabs', 'r', 'l', 'b', 'feH']].copy()

for column in data_pca:
    ret = ((data_pca[column] - data_pca[column].mean())/data_pca[column].std())
    data_pca.loc[:, column] = ret.to_numpy()
sns.scatterplot(data=data_pca, x='r', y='vabs')
```

```python
pca_vals = pca.fit_transform(data_pca.to_numpy())
pca.fit(data_pca.to_numpy())
```

```python
print(pca.explained_variance_ratio_)
```

```python
sns.scatterplot(x=pca_vals[:, 0], y=pca_vals[:, 1])
```

```python
print(df_trim.columns)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
used_columns = ['teff_gspspec', 'logg_gspspec', 'mh_gspspec', 'alphafe_gspspec', 'Jphi', 'Jz', 'JR', 'Etot', 'Lx', 'Ly', 'Lz', 'ecc', 'zmax']
```

```python
data_pca = df_trim[used_columns].copy()
print(len(data_pca))
for column in used_columns:
    data_pca = data_pca.loc[np.isfinite(data_pca[column])]
print(len(data_pca))
for column in data_pca:
    ret = ((data_pca[column] - data_pca[column].mean())/data_pca[column].std())
    data_pca.loc[:, column] = ret.to_numpy()
```

```python
pca_vals = pca.fit_transform(data_pca.to_numpy())
pca.fit(data_pca.to_numpy())
```

```python
print(pca.explained_variance_ratio_)
```

```python
plot = sns.scatterplot(x=pca_vals[:, 0], y=pca_vals[:, 1])
plot.set(xlim=(-20,20))
plot.set(ylim=(-5,35))
```
