```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_hdf('../../data/gaia/raw/common_kinematics_10kpc.h5')
print(len(df))
df.head()
df=df.loc[df['zmax']>=4000]
print(len(df))
```

```python
sns.scatterplot(data=df, x='Jphi', y='Etot', s=1)
```

```python
df = df.loc[df['JR']<0.4*1e6]
ax = sns.scatterplot(x=df['Jphi'], y=df['Jz']-df['JR'])
ax.set(xlim=(-9000, 9000), ylim=(-9000, 9000))
```

```python
indices = np.random.choice(len(df), 3000)
df_small = df.iloc[indices].copy()
```

```python
sns.scatterplot(data=df_small, x='Jphi', y='Etot', s=2)
```

```python
ax = sns.scatterplot(x=df_small['Jphi'], y=df_small['JR'])
ax.set(xlim=(-4000, 4000), ylim=(-4000, 4000))
```
