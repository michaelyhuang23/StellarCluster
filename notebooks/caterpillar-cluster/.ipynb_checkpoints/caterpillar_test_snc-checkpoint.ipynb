```python
import sys
sys.path.append('../..')

import json
import pandas as pd
import torch
from torch.optim import SGD, Adam
from torch_geometric.loader import DataLoader
import torch_geometric.transforms as T
from torch.utils.tensorboard import SummaryWriter
from tools.caterpillar_dataset import NormalCaterpillarDataset
from tools.gnn_models import GCNEdgeBased
from tools.evaluation_metric import *
from tools.cluster_functions import *
from scipy.sparse import csr_array, csr_matrix
from torch_geometric.utils import remove_self_loops
from torcheval.metrics.functional import binary_accuracy

import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
writer = SummaryWriter()
device = 'cuda' if torch.cuda.is_available() else 'cpu'

print(device)
train_dataset_ids = [1130025, 1195075, 1195448, 1232164, 1268839, 1292085,\
                    1354437, 1422331, 1422429, 1599988, 1631506, 1631582, 1725139,\
                    1725272, 196589, 264569, 447649, 5320, 581141, 581180, 795802,\
                    796175, 94638, 95289]

test_dataset_ids = [1079897, 1232423, 1599902, 196078]

val_dataset_ids = [1104787, 1387186, 388476, 65777]
```

```python
def filterer(df):
    return df.loc[df['redshiftstar']<2].copy()

feature_columns = ['estar', 'jrstar', 'jzstar', 'jphistar', 'rstar', 'vstar', 'vxstar', 'vystar', 'vzstar', 'vrstar', 'vphistar', 'phistar', 'zstar']
#feature_columns = ['estar', 'jrstar', 'jzstar', 'jphistar', 'vstar', 'vzstar', 'vrstar', 'vphistar']
#feature_columns = ['estar', 'jrstar', 'jzstar', 'jphistar']
position_columns = ['xstar', 'ystar', 'zstar']
data_transforms = T.Compose(transforms=[T.KNNGraph(k=300, force_undirected=True, loop=False), T.GDC(normalization_out='sym', self_loop_weight=None, sparsification_kwargs={'avg_degree':300, 'method':'threshold'})]) # 

test_dataset = NormalCaterpillarDataset('../../data/caterpillar', '0', feature_columns, position_columns, use_dataset_ids=test_dataset_ids, data_filter=filterer, repeat=10, label_column='cluster_id', transform=data_transforms)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)
```

```python
model = GCNEdgeBased(len(feature_columns), regularizer=0).to(device)
model.load_state_dict(torch.load('../../train_script/weights/GCNEdgeBased_model300_Aug_24/299.pth', map_location=device)['model_state_dict'])
```

```python
from scipy.sparse import csr_matrix
def evaluate(n_components, graph, model):
    graph = graph.to(device)
    with torch.no_grad():
        model.eval()
        edge_pred = model(graph)
    adj = csr_matrix((edge_pred.cpu(),graph.edge_index.cpu()), shape=(len(graph.x), len(graph.x)))    
    FX = C_Spectral(adj, n_components=n_components)
    return FX
```

```python
graph = next(iter(test_loader))
FX = evaluate(5, graph, model)
```

```python
def plot(df, clusters, fig_name, fig_path = '../../', sample_size = None):
    if sample_size is not None:
        sampled_ids = np.random.choice(len(clusters), sample_size, replace=False)
        df = df.iloc[sampled_ids]
        clusters = clusters[sampled_ids]
    clusters = [f'cluster {i}' for i in clusters]
    fig, axs = plt.subplots(ncols=3,nrows=3,figsize=(15, 15))
    axs = axs.flatten()
    sns.scatterplot(data=df, x='jphistar', y='estar', hue=clusters, style=clusters, s=10, ax=axs[0])
    sns.scatterplot(data=df, x='vrstar', y='vphistar', hue=clusters, style=clusters, s=10, ax=axs[1])
    sns.scatterplot(data=df, x='vrstar', y='vzstar', hue=clusters, style=clusters, s=10, ax=axs[2])
    J = np.sqrt(df['jphistar']**2 + df['jrstar']**2 + df['jzstar']**2)
    sns.scatterplot(x=df['jphistar']/J, y=(df['jzstar']-df['jrstar'])/J, hue=clusters, style=clusters, s=10, ax=axs[3])
    sns.scatterplot(data=df, x='jrstar', y='jphistar', hue=clusters, style=clusters, s=10, ax=axs[4])
    axs[4].set(xlim=(0, 3))
    sns.scatterplot(data=df, x='jrstar', y='jzstar', hue=clusters, style=clusters, s=10, ax=axs[5])
    axs[5].set(xlim=(0, 3))
    if 'rstar' in df.keys():
        sns.scatterplot(data=df, x='rstar', y='zstar', hue=clusters, style=clusters, s=10, ax=axs[6])
        df['phistar'] *= np.pi / np.max(df['phistar'].to_numpy())
        df['xstar'] = df['rstar'] * np.cos(df['phistar'])
        df['ystar'] = df['rstar'] * np.sin(df['phistar'])
        sns.scatterplot(data=df, x='phistar', y='rstar', hue=clusters, style=clusters, s=10, ax=axs[7])
        L = np.cross(df[['xstar', 'ystar', 'zstar']].to_numpy(), df[['vxstar', 'vystar', 'vzstar']].to_numpy())
        sns.scatterplot(x=L[:,2], y=L[:,1], hue=clusters, style=clusters, s=10, ax=axs[8])
        
    fig.savefig(os.path.join(fig_path, fig_name))
```

```python
df_x = pd.DataFrame(graph.x, columns = feature_columns)
```
