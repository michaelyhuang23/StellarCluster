```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_hdf('../data/m12i_cluster_data_v2.h5', key='star')
df = df.reset_index()
```

```python
df = df.iloc[np.random.choice(len(df), 1000)].copy()
df = df.reset_index()
df['rstar'] = np.linalg.norm([df['xstar'],df['ystar'],df['zstar']], axis=0)
sns.kdeplot(data=df, x='rstar', y='feH')
```

```python
df = pd.read_hdf('../data/m12i_cluster_data_v2.h5', key='star')
df = df.reset_index()
df = df.loc[np.isin(df['cluster_id'], [67,50,53,89])].copy()
```

```python
print(len(df))
df['cluster_id_name'] = np.array([f'cluster {id}' for id in df['cluster_id'].to_numpy()])
sns.scatterplot(data=df, x='thetaphistar', y='thetarstar', hue='cluster_id_name')
```

```python
df['cluster_id_name'] = np.array([f'cluster {id}' for id in df['cluster_id'].to_numpy()])
sns.scatterplot(data=df, x='xstar', y='ystar', hue='cluster_id_name')
```
