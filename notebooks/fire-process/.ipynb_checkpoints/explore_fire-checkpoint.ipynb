##### Compute Cluster ID

```python
df = df.loc[np.isfinite(df['subhalo_infall_mass_stars'])]
```

```python
df.sort_values('subhalo_infall_stellar_mass_stars', inplace=True, kind='stable')
df.head()
```

```python
df.sort_values('subhalo_infall_mass_stars', inplace=True, kind='stable')
df.head()
```

```python
prev_val = -1
prev_id = 0
df['cluster_id'] = 0
cluster_ids = np.zeros((len(df)), dtype=np.int32)
for i, (idx, row) in enumerate(df.iterrows()):
    if (row['subhalo_infall_mass_stars'] - prev_val)>0.1:
        prev_id+=1
        prev_val = row['subhalo_infall_mass_stars']
    cluster_ids[i] = prev_id
print(cluster_ids)
df['cluster_id'] = cluster_ids
```

```python
print(len(cluster_ids))
print(cluster_ids[:10])
df.tail()
```
