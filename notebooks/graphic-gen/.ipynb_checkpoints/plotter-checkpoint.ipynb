```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style as style
import pandas as pd
import numpy as np
```

```python
style.use('seaborn-poster') #sets the size of the charts
# style.use('ggplot')

sns.set_style('white')
sns.set(font='serif')
sns.set_style("white", {
        "font.family": "serif",
        "font.serif": ["Times", "Palatino", "serif"]
    })
sns.set_context("paper", font_scale = 1.7)

cmm = sns.color_palette("icefire")

sns.set_palette(cmm)
```

```python
def get_colors():
    return np.array([
        [0.1, 0.1, 0.1],          # black
        [0.4, 0.4, 0.4],          # very dark gray
        [0.7, 0.7, 0.7],          # dark gray
        [0.9, 0.9, 0.9],          # light gray
        [0.984375, 0.7265625, 0], # dark yellow
        [1, 1, 0.9]               # light yellow
    ])
sns.palplot(get_colors())
```

```python
df=pd.DataFrame(data={'algorithm':['SNC']*5+['HDBSCAN']*5+['GMM']*5, 
                      'number':[0.1493149411,0.1804334651,0.2847,0.26,0.35944284]+\
                      [0.06937822395,0.009842432146,0.19885714,0.0282,0.070739]+\
                      [0.0156642547,0.015555555,0.06176557116,0.06166666667,0.192173], 
                      'metric':['IoU Recall', 'IoU Precision', 'Soft Recall', 'Soft Precision', 'AMI']*3})
```

```python
plt.figure(figsize=(12,4))
fig = sns.barplot(x="metric", y="number", hue="algorithm", data=df)
fig.figure.savefig('../comparison.png',dpi=300,bbox_inches="tight")
plt.show()
```
